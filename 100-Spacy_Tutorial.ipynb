```python
print(len(text))
```

## Create Patterns


```python
#Create the EntityRuler
ruler = nlp.add_pipe("entity_ruler", after="ner")

#List of Entities and Patterns
patterns = [
                {"label": "HORMONE", "pattern": "insulin"},
    {"label": "POISON", "pattern": "sugar"},
    {"label": "POISON", "pattern": "seed oils"}
    
            ]

ruler.add_patterns(patterns)
```

```python
doc = nlp(text)
```

```python
print(len(doc))
```

```python
for token in doc[0:5]:
    print(token)
```

```python
sentence1 = list(doc.sents)[2]

print(sentence1)
print(f'Sentence count: {len(list(doc.sents))}')
```

```python
token = sentence1[7]

print(token.text)
```

```python
print(f'Enc type {token.ent_type} {token.ent_type_}')
```

```python
print(token.lemma_)
print(token.morph)
print(f'Part of speech: {token.pos_}')
print(f'Language: {token.lang_}')
```

```python
displacy.render(doc, style='ent')
```

## Word count


```python
words = [token.text for token in doc if token.is_stop != True and token.is_punct != True]
word_freq = Counter(words)
common_words = word_freq.most_common(50)
print (common_words)
```
