```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load('en_core_web_sm')
nlp.max_length = 5_000_000
```

```python
filename = './data/panama1.txt'
# filename = './data/01_star_wars.txt'
# filename = './data/king_james_bible.txt'
with open (filename, "r") as f:
    text = f.read()
```

```python
print(len(text))
```

```python
doc = nlp(text)
```

```python
print(len(doc))
```

```python
for token in doc[0:5]:
    print(token)
```

```python
sentence1 = list(doc.sents)[5]

print(sentence1)
```

```python
token = sentence1[7]

print(token.text)
```

```python
print(f'Enc type {token.ent_type} {token.ent_type_}')
```

```python
print(token.lemma_)
print(token.morph)
print(f'Part of speech: {token.pos_}')
print(f'Language: {token.lang_}')
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
displacy.render(doc, style='ent')
```
